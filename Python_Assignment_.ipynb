```python
import matplotlib.pyplot as plt
import numpy as np

def menu():
  print('''List of graphs available in this program:
        1. Scatter Plot
        2. Bar Graph
        3. Horizontal Bar Graph
        4. Histogram
        5. Pie Chart
        6. Subplots
        Enter 0 to exit''')

menu()          
Option=int(input("Please choose an option from the list: "))

while(Option!=0):
  if(Option==1):
    #scatter plot - function for displaying a piechart is plt.scatter() 
    x=[]
    print("Enter values for X-axis (seperated by \",\"): ", end="")
    for num in input(). split(","):
      x.append(int(num))
    y=[]
    print("Enter values for Y-axis (seperated by \",\"): ", end="")
    for num in input(). split(","):
      y.append(int(num))
    plt.scatter(x,y)
    plt.show()

  elif(Option==2):
    #bar graph -  function for displaying a piechart is plt.bar()
    mylabels=[]
    print("Enter lables for X-axis (seperated by \",\"): ", end="")
    for label in input(). split(","):
      mylabels.append(str(label))
    y=[]
    print("Enter values for Y-axis (seperated by \",\"): ", end="")
    for num in input(). split(","):
      y.append(int(num))
    plt.bar(mylabels,y)
    plt.show()

  elif(Option==3):
    #horizontal bar graph - function for displaying a piechart is plt.barh()
    mylabels=[]
    print("Enter labels for X-axis (seperated by \",\"): ", end="")
    for label in input(). split(","):
      mylabels.append(str(label))
    y=[]
    print("Enter values for Y-axis (seperated by \",\"): ", end="")
    for num in input(). split(","):
      y.append(int(num))
    plt.barh(mylabels,y)
    plt.show()

  elif(Option==4):
    #histogram - function for displaying a piechart is plt.hist()
    x=[]
    print("This option prints the histogram by importing random values from numpy")
    x=np.random.normal(170, 10, 250)
    plt.hist(x)
    plt.show()

  elif(Option==5):
    #piechart - function for displaying a piechart is plt.pie()
    mylabels=[]
    print("Enter labels for piechart (seperated by \",\"): ", end="")
    for label in input(). split(","):
      mylabels.append(str(label))
    y=[]
    print("Enter values for Y-axis (seperated by \",\"): ", end="")
    for num in input(). split(","):
      y.append(int(num))
    plt.pie(y,labels= mylabels,startangle=90)
    plt.show()

  elif(Option==6):
    #subplots - used to divide the visualization area into sub graphs, function used is plt.subplot()
    print("In this option we will be dividing the visualization area into 6 sub plots, please enter the required values for X-axis and Y-axis.")
    print("\n")
    
    #graph1
    print("Graph 1")
    x=[]
    print("Enter values for X-axis (seperated by \",\"): ", end="")
    for num in input(). split(","):
      x.append(int(num))
    y=[]
    print("Enter values for Y-axis (seperated by \",\"): ", end="")
    for num in input(). split(","):
      y.append(int(num))
    plt.subplot(2,3,1)      #dividing visualization area into 2 rows and 3 columns
    plt.plot(x,y)
    print("\n")

    #graph2
    print("Graph 2")
    x=[]
    print("Enter values for X-axis (seperated by \",\"): ", end="")
    for num in input(). split(","):
      x.append(int(num))
    y=[]
    print("Enter values for Y-axis (seperated by \",\"): ", end="")
    for num in input(). split(","):
      y.append(int(num))
    plt.subplot(2,3,2)
    plt.plot(x,y)
    print("\n")

    #graph3
    print("Graph 3")
    x=[]
    print("Enter values for X-axis (seperated by \",\"): ", end="")
    for num in input(). split(","):
      x.append(int(num))
    y=[]
    print("Enter values for Y-axis (seperated by \",\"): ", end="")
    for num in input(). split(","):
      y.append(int(num))
    plt.subplot(2,3,3)
    plt.plot(x,y)
    print("\n")

    #graph4
    print("Graph 4")
    x=[]
    print("Enter values for X-axis (seperated by \",\"): ", end="")
    for num in input(). split(","):
      x.append(int(num))
    y=[]
    print("Enter values for Y-axis (seperated by \",\"): ", end="")
    for num in input(). split(","):
      y.append(int(num))
    plt.subplot(2,3,4)
    plt.plot(x,y)
    print("\n")

    #graph5
    print("Graph 5")
    x=[]
    print("Enter values for X-axis (seperated by \",\"): ", end="")
    for num in input(). split(","):
      x.append(int(num))
    y=[]
    print("Enter values for Y-axis (seperated by \",\"): ", end="")
    for num in input(). split(","):
      y.append(int(num))
    plt.subplot(2,3,5)
    plt.plot(x,y)
    print("\n")
    
    #graph6
    print("Graph 6")
    x=[]
    print("Enter values for X-axis (seperated by \",\"): ", end="")
    for num in input(). split(","):
      x.append(int(num))
    y=[]
    print("Enter values for Y-axis (seperated by \",\"): ", end="")
    for num in input(). split(","):
      y.append(int(num))
    plt.subplot(2,3,6)
    plt.plot(x,y)

    plt.show()

  else:
    print("Sorry, this option is out of list.")

  print()  
  menu()
  Option=int(input("Please choose an option from the list: "))
```
